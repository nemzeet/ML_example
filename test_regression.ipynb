```python
#  -> train/valid, test
from sklearn.model_selection import train_test_split

tr_val_X, test_X, tr_val_Y, test_Y = train_test_split(
    input_data[X_column], # X  
    input_data[Y_column], # Y  
    test_size = 0.2,
    shuffle = True,
    random_state = 42 #    sampling
    ) 
```

```python
tr_val_data = pd.concat([tr_val_X, tr_val_Y], axis = 1)
tr_val_data
```

```python
#  -> train/valid, test
from sklearn.model_selection import train_test_split

train_X, valid_X, train_Y, valid_Y = train_test_split(
    tr_val_data[X_column], # X  
    tr_val_data[Y_column], # Y  
    test_size = 0.2,
    shuffle = True,
    random_state = 42 #    sampling
    ) 
```

```python
print(len(train_X))
print(len(valid_X))
print(len(test_X))
len(train_X) + len(valid_X) + len(test_X) # 20
```

```python
# 4,  
```

```python
# Decision Tree, Random Forest, XGB, LGBM, CATBOOST
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor()
model_rf.fit(train_X, train_Y) # train_X   train_Y   ..
```

```python
display(test_X, test_Y) #    
```

```python
true_y = test_Y
true_y
```

```python
# predicton ()
pred_y = model_rf.predict(test_X)
pred_y
```

```python
df_result = pd.DataFrame(list(zip(test_Y, pred_y)), columns=['true_Y', 'pred_y'])
df_result
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(true_y, pred_y)
mse
```

```python
# RMSE root mean
rmse = np.sqrt(mse)
mse
```

```python
from xgboost import XGBRegressor

model_xgb = XGBRegressor()
model_xgb.fit(train_X, train_Y, eval_set=[(valid_X, valid_Y)])
```

```python
# prediction ()
xgb_pred_y = model_xgb.predict(test_X)
xgb_pred_y
```
