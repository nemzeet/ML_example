```python
# 모델에 사용할 컬럼 결정 (column, feature) 결정
input_data = df[X_column + [Y_column]].copy() # .copy() : Memory input_data 메모리 설정
input_data
```

```python
input_data['Embarked'] = input_data['Embarked'].replace({'S': 0, 'C':1 })
input_data['Embarked']
```

```python
# 모델 학습
# 1. train data 학습
#   - train
#   - validation data 학습 하는 도중에 실행하는 성능 평가
# 2, test data 모델 성능 평가
```

```python
# 전체 데이터 20개
# 16개 -> train (8), validation (6) ---> python의 슬라이싱
# 4개 -> test
```

```python
# 데이터 -> train/valid, test
from sklearn.model_selection import train_test_split

tr_val_X, test_X, tr_val_Y, test_Y = train_test_split(
    input_data[X_column], # X 입력 데이터
    input_data[Y_column], # Y 입력 데이터
    test_size = 0.2,
    shuffle = True,
    random_state = 42 # 항상 동일한 변수로 sampling
    ) 
```

```python
tr_val_data = pd.concat([tr_val_X, tr_val_Y], axis = 1)
tr_val_data
```

```python
# 데이터 -> train/valid, test
from sklearn.model_selection import train_test_split

train_X, valid_X, train_Y, valid_Y = train_test_split(
    tr_val_data[X_column], # X 입력 데이터
    tr_val_data[Y_column], # Y 입력 데이터
    test_size = 0.2,
    shuffle = True,
    random_state = 42 # 항상 동일한 변수로 sampling
    ) 
```

```python
print(len(train_X))
print(len(valid_X))
print(len(test_X))
len(train_X) + len(valid_X) + len(test_X) # 20
```

```python
# 4, 모델 학습
```

```python
# Decision Tree, Random Forest, XGB, LGBM, CATBOOST
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor()
model_rf.fit(train_X, train_Y) # train_X를 넣었을 때 train_Y라는 결과를 도출하게 해줌..
```

```python
display(test_X, test_Y) # 여러 데이터프레임을 한번에 보여줌
```

```python
true_y = test_Y
true_y
```

```python
# predicton (예측)
pred_y = model_rf.predict(test_X)
pred_y
```

```python
df_result = pd.DataFrame(list(zip(test_Y, pred_y)), columns=['true_Y', 'pred_y'])
df_result
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(true_y, pred_y)
mse
```

```python
# RMSE root mean
rmse = np.sqrt(mse)
mse
```

```python
from xgboost import XGBRegressor

model_xgb = XGBRegressor()
model_xgb.fit(train_X, train_Y, eval_set=[(valid_X, valid_Y)])
```

```python
# prediction (예측)
xgb_pred_y = model_xgb.predict(test_X)
xgb_pred_y
```
